```python
#на основе этих данных была создана таблица в эксель для создания круговой диаграммы
df=pd.read_excel('/content/Pie_diagram.xlsx')
df
```

```python
labels = ["Sword", "Arch", "Catalyst", "Spear"]
values = [20, 8, 7, 6]
colors = ['yellow','green','red','blue']
explode = [0.3,0,0,0]
plt.title('A Pie Chart')
plt.pie(values,labels=labels,colors=colors,explode=explode,shadow=True,autopct='%1.1f%%',startangle=180)
plt.axis('equal')
plt.show()
```
